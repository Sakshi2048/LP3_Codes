```python
# convert ORDERDATE to numeric attribute
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
data['ORDERDATE'] = data['ORDERDATE'].astype('int64')

data['STATUS'] = data['STATUS'].replace('Shipped',1)
data['STATUS'] = data['STATUS'].replace('Resolved', 2)
data['STATUS'] = data['STATUS'].replace('Cancelled', 3)
data['STATUS'] = data['STATUS'].replace('On Hold', 4)
data['STATUS'] = data['STATUS'].replace('Disputed', 5)
data['STATUS'] = data['STATUS'].replace('In Process', 6)


data = data.drop(columns=['PRODUCTCODE'])

data = data.drop(columns=['ADDRESSLINE1', 'ADDRESSLINE2', 'POSTALCODE', 'TERRITORY', 'CITY', 'STATE', 'COUNTRY'])

data = data.drop(columns=['CUSTOMERNAME', 'CONTACTLASTNAME','CONTACTFIRSTNAME', 'DEALSIZE', 'PHONE'])

data = pd.get_dummies(data, columns=['PRODUCTLINE'])
```

```python
data.describe()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dataset = data.values

WCSS = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(dataset)
    WCSS.append(kmeans.inertia_)

plt.plot(range(1,20), WCSS)

kmeans = KMeans(n_clusters=3)
kmeans.fit(dataset)
labels = kmeans.predict(dataset)
centroids = kmeans.cluster_centers_
```
